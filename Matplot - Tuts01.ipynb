# Axes labels, Legend, Grid

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.plot(days, max_t) #x,y
```

```python
plt.plot(days, max_t) #x,y
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

```python
plt.xlabel('Days')
plt.ylabel('Temprature')
plt.title("Weather")
plt.plot(days, max_t) #x,y
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

This is still not perfect 

```python
plt.xlabel('Days')
plt.ylabel('Temprature')
plt.title("Weather")
plt.plot(days, max_t, label='Max') #x,y
plt.plot(days, min_t, label='Min')
plt.plot(days, avg_t, label='Avg')

# plt.legend()
plt.legend(loc='best', fontsize='large')
plt.grid()
```
