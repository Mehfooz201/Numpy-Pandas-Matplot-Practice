# Pandas - Data Manipulation

    - Handle the Missing Values (using fillna=0 or something else)
    - Interpolate to make a guess on missing values using interpolation
    - Dropna to dr0p rows with missing values

```python
import pandas as pd

#Inplace=True
#Parse_dates=['day']
#fillna = 0
```

```python
df = pd.read_csv('D:\\4. Programming, AI, ML\\04 Machine Learning\\Numpy and Pandas\\wheather_data_ms_data.csv', parse_dates=['day'])
df.set_index('day', inplace=True)
df
```

```python
df['temperature'].min()
```

```python
type(df.day[0])
```

```python
#fillna to fill NaN values with giving desire argumement
#fillna = 0
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event' : 'No Event',
    
})

new_df
```

# FFIL Method 
 - Previous Values (Previous Day)

```python
new_df = df.fillna(method='ffill')
new_df
```

```python
new_df = df.fillna(method='bfill')
new_df
```

```python
new_df = df.fillna(method='bfill', axis='columns')
new_df
```

```python
new_df = df.fillna(method='ffill')
new_df
```

# Interpolate
    - Guesss the values

```python
new_df1 = df.interpolate(method='time')
new_df1
```

# Replace the Value

```python
import numpy as np
```

```python
df4 = pd.read_csv('wheather_data_updated.csv', parse_dates=['day'])
df4
```

```python
new_df = df4.replace([-99999, -88888], np.NaN)
new_df
```

```python
new_df = df4.replace({
    'temperature' : -99999,
    'windspeed' : [-99999, -88888],
    'event' : 0
}, np.NaN)

new_df
```

```python
new_df = df4.replace({
    -99999 : np.NaN,
    -88888 : np.NaN,
    'No Event' : 'Sunny',
    
})

new_df
```

# Regex => Regular Expression
     - Used for detect patern

```python
new_df = df4.replace('[A-Za-z]', '', regex=True)
new_df
```
