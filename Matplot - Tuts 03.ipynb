# Matplotlib 
    - Histograms
    - PyChart


```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Histograms

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar)
```

# Bins

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, bins=3)
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, bins=3, rwidth=0.95)
```

# Blood sample for men and women

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 110, 100]

plt.xlabel('Sugar Range')
plt.ylabel('Total Number of Patient')
plt.title('Blood Sugar Analysis')

plt.hist([blood_sugar_men, blood_sugar_women], bins=3, rwidth=0.95, label=['men', 'women'])

plt.legend()
```

# Horizontal

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 110, 100]

plt.xlabel('Sugar Range')
plt.ylabel('Total Number of Patient')
plt.title('Blood Sugar Analysis')

plt.hist([blood_sugar_men, blood_sugar_women], bins=3, rwidth=0.95, label=['men', 'women'], orientation='horizontal')

plt.legend()
```

# PyChart

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ['Home Rent', 'Foood', 'Phone/Internet Bill', 'Car', 'Other Utilities']
```

```python
plt.pie(exp_vals, labels=exp_labels)
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels)
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.1f%%')
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.1f%%', explode=[0,0.2, 0, 0, 0])
plt.show()
```

# How to save the Pychart in file

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.1f%%', explode=[0,0.2, 0, 0, 0])

plt.savefig('piechart.jpg', bbox_inches='tight')
```
