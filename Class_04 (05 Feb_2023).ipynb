# Combining and Reshaping Data

# Merging and Joining Data Frame

```python
import pandas as pd
import numpy as np
import datetime
```

```python
customers = {'customer_id' : [1234, 1235],
             'Name' : ['Mike', 'Maria'],
             'Address' : ['acb23-1', 'xyz-3'],            
            }

customers = pd.DataFrame(customers)
```

```python
customers.set_index('customer_id')
```

```python
customers.info()
```

```python
orders = {'customer_id' : [1234, 1235, 1234, 1234, 1235, 1267, 1237, 1890],
          'order_id' : [11,22,33,44,55,66,77,88],
          'order_date':[datetime.date(2023,2,1), datetime.date(2023,1,1), 
                        datetime.date(2023,2,2),datetime.date(2023,1,1),
                        datetime.date(2023,2,3),datetime.date(2023,1,2),
                        datetime.date(2023,2,4),datetime.date(2023,2,1)],
         }

orders = pd.DataFrame(orders)
orders
```

```python
orders.info()
```

```python
merged_data = pd.merge(customers, orders)
merged_data
```

```python
merged_data = merged_data[['Name', 'order_id', 'order_date']]
merged_data
```

# Merge
    - Inner
    - Outer
    - Left
    - Right
    
    - Concatenate => Number or Row and Column will be same
    - Merging => Befault check which column overlap 

```python
merged_data.sort_values(by='customer_id', ascending=False)
```

```python
merged_data = pd.merge(orders, customers, how='inner')
merged_data
```

```python
merged_data = pd.merge(orders, customers, how='outer')
merged_data #union
```

```python
merged_data = pd.merge(customers, orders, how='left')
merged_data #union
```

```python
merged_data = pd.merge(customers, orders, how='right')
merged_data #union
```
